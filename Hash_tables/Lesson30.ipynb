```python
class HashTable:
    def __init__(self, length):
        self.length = length
        self.table = [None] * length

    def hash_function(self, key):
        return key % self.length

    def add_item(self, key, value):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = (key, value)
                return
            index = (index + 1) % self.length
        self.table[index] = (key, value)

    def get_item(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.length
        return None

    def delete_item(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = None
                return
            index = (index + 1) % self.length

    def __str__(self):
        result = ''
        for key_value in self.table:
            if key_value is not None:
                result += str(key_value[0]) + ' -> ' + str(key_value[1]) + ', '
        return '{' + result[:-2] + '}'
```

```python
def hash_password(password):
    return sum(map(ord, password))

def register(users):
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    if username in users:
        print("This username is already taken. Please choose another one.")
    else:
        users[username] = hash_password(password)
        print("Registration successful.")

def login(users):
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    if (username in users) and (users[username] == hash_password(password)):
        print("Login successful.")
    else:
        print("Login failed. Please check your username and password.")

def main():
    users = {}
    while True:
        action = input("Do you want to register or login? (R/L): ")
        if action.lower() == 'r':
            register(users)
        elif action.lower() == 'l':
            login(users)
        else:
            print("Invalid action. Please enter either R for register or L for login.")
            continue

main()
```
